##**Exploratory Data Analysis**

###<b>Order Priority</b>

```python
#Count of orders based on Order priority
df['Order Priority'].value_counts()
```

```python
df['Order Priority'].unique()
```

```python
df['Order Priority'] = df['Order Priority'].replace('Critical ','Critical')
```

```python
df['Order Priority'].value_counts()
```

```python
#Plotting a countplot
sns.set(rc={'figure.figsize':(6,4)})
ax = sns.countplot(x='Order Priority', data = df)

plt.ylabel('Number of orders')
for bars in ax.containers:
  ax.bar_label(bars)
```

*Above plot indicates that number of orders is highest in the High order priority segment followed by low order priority.*

###<b>Shipping Mode</b>

```python
#Number of orders Shipped through different shipping mode
df['Ship Mode'].value_counts()
```

```python
#Percentage orders shipped through different modes

#Plotting a pie chart
x = df['Ship Mode'].value_counts().index
y = df['Ship Mode'].value_counts().values

plt.figure(figsize=(4,4))
plt.pie(y,labels=x,autopct='%0.1f%%',startangle=60)
plt.legend(loc=2)
plt.show()
```

*Above pie chart shows us that most of the orders are shipped through Regular Air Shipping Mode.*

```python
#Number of orders of different Product Category shipped through different modes

sns.set(rc={'figure.figsize':(5,4)})
ax = sns.countplot(x='Ship Mode',data=df,hue='Product Category')
plt.title('Number of orders of different Product Category \nshipped through different modes\n')
plt.ylabel('Number of orders')
for bars in ax.containers:
  ax.bar_label(bars)
```

*From the above graph we can see that most of the products in each category are shipped through Regular Air service.*

###<b>Customer Segment</b>

```python
#Number of order placed by different Customer Segment

#Plotting a countplot
sns.set(rc={'figure.figsize':(7,4)})
ax = sns.countplot(x='Customer Segment', data = df)

plt.ylabel('Number of orders')
for bars in ax.containers:
  ax.bar_label(bars)
```

*From above graph we can see that most of the orders are placed by Corporate Customer Segment followed by Home Office segment and Consumer Segment has the least number of orders.*

###<b>Product category</b>

```python
#Number of orders based on different product category

#Plotting a countplot
sns.set(rc={'figure.figsize':(5,4)})
ax = sns.countplot(x='Product Category', data = df)

plt.ylabel('Number of orders')
for bars in ax.containers:
  ax.bar_label(bars)
```

*From the above graph we can see most peoples are likely to buy Office Supplies.*

```python
#Plotting countplot

sns.set(rc={'figure.figsize':(15,7)})
ax = sns.countplot(x='Product Category', data = df[df['Product Category']=='Office Supplies'], hue='Product Sub-Category')
plt.title('Number of orders for different sub-categories in Office Supplies')
plt.ylabel('Number of orders')
for bars in ax.containers:
  ax.bar_label(bars)
```

*Above graph shows us the most ordered Product sub-category in Office Supplies is Paper followed by Binders and Binder Accessories.*

```python
#Plotting countplot
sns.set(rc={'figure.figsize':(7,5)})
ax = sns.countplot(x='Product Category', data = df[df['Product Category']=='Technology'], hue='Product Sub-Category')
plt.title('Number of orders for different sub-categories in Technology')
plt.ylabel('Number of orders')
for bars in ax.containers:
  ax.bar_label(bars)
```

*From the above plot we can see that the most ordered Product sub-category in Technology Category is Telephones and Communication followed by Computer Peripherals.*

```python
#Plotting Countplot

sns.set(rc={'figure.figsize':(6,4)})
ax = sns.countplot(x='Product Category', data = df[df['Product Category']=='Furniture'], hue='Product Sub-Category')
plt.title('Number of orders for different sub-categories in Furniture')
plt.ylabel('Number of orders')
for bars in ax.containers:
  ax.bar_label(bars)
```

*Above graph shows us the most ordered Product sub-category in Furniture Category is Office Furnishings followed by Chairs & Chairmats.*

###**Order Year**

```python
#Creating a new column order year
df['Order Year'] = df['Order Date'].dt.year
```

```python
df.head()
```

```python
#Number of order placed in different years
yearly_orders=df['Order Year'].value_counts().sort_index()
```

```python
#Creating a plot object
plt.figure(figsize=(3,4))
yearly_orders.plot()

plt.title('Number of orders placed in different years')
plt.xlabel('Order Year')
plt.ylabel('Number of orders')
plt.show()
```

*Above plot indicates the growth in number of orders from year 2010 to 2013.*

###**Yearly sales growth**

```python
yearly_sales_growth = df.groupby('Order Year')['Sales'].sum()
```

```python
yearly_sales_growth
```

```python
plt.figure(figsize=(3,4))
yearly_sales_growth.plot()

plt.title('Yearly Sales')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()
```

*From the above graph we can see the growth in sales from year 2010 to 2013.*

###**Profit**

```python
#Profit made per product category
plt.figure(figsize=(5,4))

plt.figure(figsize=(5,4))
sns.set(rc={'figure.figsize':(5,4)})
ax = sns.barplot(x='Product Category',y='Profit',data=df,estimator=sum)
plt.title('Profit by Product Category')
for bars in ax.containers:
  ax.bar_label(bars)
```

*From the above graph we can see that the profit made by different product category where Technology category has the highest profit.*

```python
#Plotting a barplot
plt.figure(figsize=(5,4))
sns.set(rc={'figure.figsize':(5,4)})
ax = sns.barplot(data=df, x='Product Category', y='Product Base Margin', estimator=sum)
plt.title("Product Base Margin by Product Category\n")
for bars in ax.containers:
  ax.bar_label(bars)
```

*Above graph indicates that Office Supplies has the highest product margin.*

###**Region**

```python
#Orders received from different regions

x = df['Region'].value_counts().index
y = df['Region'].value_counts().values

plt.figure(figsize=(6,5))
plt.pie(y,labels=x,autopct='%0.1f%%',startangle=120)
plt.legend(loc=2)
plt.show()
```

*From the above figure we can see almost 1/3rd of orders are placed from the central region.*

```python
states_orders=df['State or Province'].value_counts()
```

```python
plt.figure(figsize=(10,6))
states_orders.plot(kind='bar')
plt.xlabel('State or Province')
plt.ylabel('Number of orders')
plt.title('Number of orders received from different states')
plt.show()
```

*Looking at the above plot it is clear that the top 5 States with most numbers of orders are California, Texas, Illinois, New York and Florida.*
